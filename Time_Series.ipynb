Shift function
Shift function means shifting the column down by one month to project stock price for the next month. 

```python
amzn_monthly_return = amzn_mothend/amzn_mothend.shift(1)-1
```

```python
amzn_monthly_return.head()
```

```python
amzn_monthly_return = amzn_monthly_return.dropna()
```

```python
amzn_monthly = pd.plotting.autocorrelation_plot(amzn_monthly_return)
```

```python
mmm_monthly_return = mmm_mothend/mmm_mothend.shift(1)-1
```

```python
mmm_monthly_return.head()
```

```python
mmm_monthly_return = mmm_monthly_return.dropna()
```

```python
mmm_monthly = pd.plotting.autocorrelation_plot(mmm_monthly_return)
```

```python
fb_monthly_return = fb_mothend/fb_mothend.shift(1)-1
```

```python
fb_monthly_return.head()
```

```python
fb_monthly_return = fb_monthly_return.dropna()
```

```python
fb_monthly = pd.plotting.autocorrelation_plot(fb_monthly_return)
```

```python
ibm_monthly_return = ibm_mothend/ibm_mothend.shift(1)-1
```

```python
ibm_monthly_return.head()
```

```python
ibm_monthly_return = ibm_monthly_return.dropna()
```

```python
ibm_monthly = pd.plotting.autocorrelation_plot(ibm_monthly_return)
```

Return graphs of all the four stock (Amazon,MMM,IBM,fb) are showing the similar trend. Basically it is showing no change at all in the monthly cycle as all the points are more or less around the zero mark. In conclusion no significant trend as well as corrleation is observed in the data plotted using the shift funtion.
